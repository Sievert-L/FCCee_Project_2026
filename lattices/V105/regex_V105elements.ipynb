```python
# Overview of regex for LCC V105

# The FCC-ee has 8 arcs and 8 straight sections, of which 4 are straight sections with IPs (interaction regions or IRs) and 4 are straight sections without IPs (technical regions or TRs). 

# Abbreviations:
# IP = interaction point
# FD = Final Doublet
# FF = Final Focussing
# DS = Dispersion Suppression
# CS = Crab Sextupole

##############
# ARC ELEMENTS
##############
# arc dipoles: pattern='dl1a'
# arc quadrupoles: pattern=['qf2a', 'qf3a', 'qd1a']
# arc sextupoles: pattern='s[df][12]a'

################
# OTHER ELEMENTS
################
# FD quads: pattern=['qd0a', 'qd0b', 'qd0cr', 'qd0cl', 'qf1a', 'qf1b', 'qf1cr', 'qf1cl', 'qf1dr', 'qf1dl'] # final doublet quads in the 4 straight sections w/ IP
# FD-->CS quads: pattern=['q[xy][0-4][lr]', 'q[fd][2-9][lr]', 'q[fd]1[0-9][lr]', 'q[fd]20[lr]', '(?!qd0a|qd0b|qd0cr|qd0cl|qf1a|qf1b|qf1cr|qf1cl|qf1dr|qf1dl)q[fd][01][abcd][lr]'] # quads between final doublet and crab sextupoles in the 4 straight sections w/ IP
# CS-->arc quads and non-IP quads: pattern=['q[fd]5f', 'q[fd][7-9]m', 'q[fd][6-9][cfdij]', 'q[fd]1[0-8][cfdijm]'] # quads between crab sextupoles and arc in the 4 straight sections w/ IP as well as quads in the 4 straight sections w/o IP
# ds quads: pattern=['q[fd][01234]f', '(?!qd0c[lr]|qf1c[lr]|qf1d[lr])q[fd][012345][cdij]', 'q[fd][0123456]m'] # note: there are 6-8 quads inside each of the 16 ds regions matching the arc quad regex; the remaining quads inside the ds regions are accessed via the ds_quads regex (but ds_quads to be considered as arc elements for all practical purposes)

# FD-->CS sextupoles: pattern=['s[fd][mxy][12][lr]', 'scrab'] # IR sextupoles, i.e. sextupoles between final doublet and crab sextupoles in the 4 straight sections w/ IP as well as the crab sextupoles themselves
# DS sextupoles: pattern=['sf3m', 'sf4i', 's[df][12]bf', 's[fd][123][fdci]', 'sf4[cdf]'] # note there are 4 sextupoles inside each of the 16 ds regions matching the arc sextupole regex; the remaining sextupoles inside the ds regions are accessed via the ds_sext regex (but ds_sext to be considered as arc elements here for all practical purposes) 

# IP dipoles: pattern='b[0-7]' # dipoles located inside the 4 straight sections w/ IP
# Non-IP dipoles: pattern=['vsep[12]', 'dog[lr]_coll', 'dog[lr]_rf', 'dog[lr]_diag', 'dog[lr]_inj'] # dipoles located inside the 4 straight sections w/o IP
# RF dipoles: pattern=['dl[089][lr]_rf', 'ds[123][lr]_rf'] # dipoles located close to the RF cavity but NOT located inside one of the 8 straight sections and not accessed via the arc dipole regex
# other dipoles: pattern='d[ifs][12]a' # dipoles located inside the ds regions (apart from 7 dipoles which seem to be located inside the straight sections: df1a, di1a, di2a, di2a.0, di1a.0, ds1a.13, ds1a.14, according to the markers defined in the V106 lattice); to be treated as arc dipoles for now

########################
# Note on regex patterns:
########################
# Note that these regex patterns are tailored towards our _find_elements function with the following structure:
# Note in particular the .* in the pattern regex
# tt_element = sum((line_table.rows[f'{p}.*'] for p in patterns), line_table.rows[[]])

# Note also that the V106 regex patterns are different from the V105 patterns for the non-IP dipoles and other dipoles in particular
```
